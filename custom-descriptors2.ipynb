## Учебник Воробьёва

Как устроен вирус:
Зачастую он сферической формы. В центре нуклеокапсид - хранит геном, состоит из 7(C) или 8 секций (A, B). В основном состоит из нуклеопротеина(NP) а также белков полимеразного состава. Нуклеокапсид окружён окружён слоем матриксных и мембранных белков (M), которые участвуют в сборке вирусной частицы. Поверх этих структур располагается суперкапсид - наружная липопротеиновая оболочка ЗА СЧЁТ КОТОРОЙ вирусы гриппа чувствительный к ЭФИРАМ. Липопротеиновая оболочка имеет клеточное происхождение. Она несёт насёт на своей поверхности шипики - выросты длиной около 10 нм. Шипики образованы двумя сложными белками гликопротеином и нейроминидазой. колчество гамагглютинина(H) в 5 раз больше нейроминидазы(N). У C вирусов нет N. H и N белки кодируются вирусным геномом и в процессе репродукции вирусов встраиваются в мембрану клетки хозяина. Шипы нейроминидазы - тример, состоят из 4 молекул белка. На поверхности всех гликопротеинов есть специальные области для связывания с рецепторами. Гемагглютенины вируса гриппа связываются с рецепторами на чувствительных клетках, а затем нейроменидаза их модифицирует и вирус проникает в клетку методом эндоцитоза. Также нейроминидаза участвует в выходе из клетки новых вирионов. Снижает ВЯЗКОСТЬ секретов  

#### Способы получить гликопротеины. 
1) Вирусную частицу обрабатывают ферментами, например бромелином \
2) Обработка детергенами, разрушающими липидный бислой, молекулы высвобождаются невредимыми

#### Стратегия генома

Специфический рецептор вируса гриппа - сиаловая кислота. Когда вирус проник в клетку происходит депротеинезация и сердцевина вириона транспортируется к ядрку клетки. На ядерной оболочке происходит завершение депротеинезации т. е удаление M белка. Чтобы вирус функционировал нужна полимераза.

```python
class MolDataset(Dataset):
    def __init__(self, x, y):
        super().__init__()
        self.X = torch.tensor(x.values, dtype=torch.float32)
        self.Y = torch.tensor(y.values, dtype=torch.float32)
        self.len_ = x.shape[0]

    def __len__(self):
        return self.len_

    def __getitem__(self, index):
        return (self.X[index], self.Y[index])
```

```python
class MLP(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, 4096*2),
            nn.LeakyReLU(),
            nn.Dropout(0.3),
            nn.Linear(4096*2, 2048*2),
            nn.LeakyReLU(),
            nn.Dropout(0.1),
            nn.Linear(2048*2, 512*2),
            nn.LeakyReLU(),
            nn.Linear(512*2, 256*2),
            nn.LeakyReLU(),
            nn.Dropout(0.1),
            nn.Linear(256*2, 1),
        )
    def forward(self, x):
        return self.model(x)
```

## CNN + bidirectional GRU

```python
import torch
import torch.nn as nn

class CNN_GRU_Model(nn.Module):
    def __init__(self, input_size):
        super(CNN_GRU_Model, self).__init__()
        cnn_in_channels = 1
        cnn_out_channels = 128
        cnn_kernel_size = 3
        gru_hidden_size = 256
        num_layers = 16
        output_size = 1
        # CNN layers
        self.cnn = nn.Sequential(
            nn.Conv1d(cnn_in_channels, cnn_out_channels, cnn_kernel_size),
            nn.ReLU(),
            nn.BatchNorm1d(cnn_out_channels),
            nn.MaxPool1d(kernel_size=2)
        )
        
        # GRU layers
        self.gru = nn.GRU(input_size=cnn_out_channels, hidden_size=gru_hidden_size, num_layers=num_layers, batch_first=True, bidirectional=True)
        
        # Fully connected layer
        self.fc = nn.Linear(gru_hidden_size * 2, output_size)  # Multiply by 2 for bidirectional GRU
        
    def forward(self, x):
        x = x.view(x.size(0), 1, -1)
        # CNN forward pass
        x = self.cnn(x)
        
        # Reshape output for GRU
        x = x.permute(0, 2, 1)  # Reshape to (batch_size, seq_len, input_size) for GRU
        
        # GRU forward pass
        gru_out, _ = self.gru(x)
        
        # Take only the last time step output from both directions
        gru_out = torch.cat((gru_out[:, -1, :gru_out.size(2)//2], gru_out[:, 0, gru_out.size(2)//2:]), dim=1)
        
        # Fully connected layer
        output = self.fc(gru_out)
        return output
```

```python
class CustomTransformer(nn.Module):
    def __init__(self, input_size):
        super(CustomTransformer, self).__init__()
        
        num_heads = 1
        hidden_size = 512
        num_layers = 5
        output_size = 1
        dropout=0.3
        
        self.transformer = nn.Transformer(
            d_model=input_size,
            nhead=1,
            num_encoder_layers=num_layers,
            num_decoder_layers=num_layers,
            dim_feedforward=hidden_size,
            dropout=dropout
        )
        
        self.fc = nn.Linear(input_size, output_size)
        
        self.init_weights()

    def init_weights(self):
        initrange = 1
        self.fc.bias.data.zero_()
        self.fc.weight.data.uniform_(-initrange, initrange)

    def forward(self, src):
        # Преобразуем размерность входных данных для совместимости с трансформером
        src = src.unsqueeze(0)  # Добавляем размерность seq_len=1
        
        # Передаем входные данные через трансформер
        output = self.transformer(src, src)
        
        # Выбираем последний элемент из каждой последовательности
        output = output[-1, :, :]
        
        # Применяем полносвязный слой для получения предсказаний
        predictions = self.fc(output)
        
        return predictions.squeeze()
```

```python
import copy

class DopTrainer:
    def __init__(
        self,
        loss_func,
        optim = torch.optim.Adam,
        model = MLP,
        device: str = "cpu",
        epoch_amount:int=1000,
        batch_size:int = 50,
        learning_rate: float = 1e-4,
        permutate: bool = True,
        best_score = 0
    ) -> None:
        self.device = torch.device(device)
        self.epoch_amount = epoch_amount
        self.loss = loss_func
        self.dataset = MolDataset
        self.net = model
        self.optim = optim
        self.lr = learning_rate
        self.shufle = permutate
        self.batch_size = batch_size
        self.model = model

        self.train_loss = []
        self.train_accuracy = []

        self.validation_loss = []
        self.validation_accuracy = []

        self.best_model = model,
        self.best_score = best_score

    def predict(self, X):
        X = X.to(clf.device)
        Model = self.best_model
        Model.to(self.device)
        Model.eval()
        
        return Model(X)

    def fit(
            self,
            x_train:pd.core.frame.DataFrame,
            y_train:pd.core.frame.DataFrame,
            x_val:pd.core.frame.DataFrame,
            y_val:pd.core.frame.DataFrame
        ):
        log_step = 100

        Model = self.model

        Model.to(device=self.device)

        opt = self.optim(Model.parameters(), lr=self.lr)

        train_ds = MolDataset(x_train, y_train)
        val_ds = MolDataset(x_val, y_val)

        train_dl = DataLoader(train_ds, batch_size=self.batch_size)
        val_dl = DataLoader(val_ds, batch_size=self.batch_size, shuffle=False)
        
        sum_val = 0
        n_samples = 0
        for i, (data, classes) in enumerate(val_dl):
            sum_val += torch.sum(classes).item()
            n_samples += classes.size(0)
        
        val_mean = sum_val / n_samples
        
        for epoch in range(self.epoch_amount):
            loss_tr, loss_v = 0, 0
            start = dt.datetime.now()
            
            Model.train()
            i = 0
            for i, (data, classes) in enumerate(train_dl):
                
                data = data.to(self.device)
                classes = classes.to(self.device)
                opt.zero_grad()
                
                y_pred = Model(data)
                if (i % log_step == 0):
                    print("samples handled: ", i*batch_size)
                curr_loss = self.loss(y_pred.view(-1), classes.view(-1))

                curr_loss.backward()
                opt.step()
                loss_tr += curr_loss
            self.train_loss.append(loss_tr.item() / len(train_dl))
            
            sse = 0
            sst = 0
            Model.eval()
            with torch.no_grad():
                for i, (data, classes) in enumerate(val_dl):
                    data = data.to(self.device)
                    classes = classes.to(self.device)
                    y_pred = Model(data)
                    sse += torch.sum((y_pred.view(-1) - classes) ** 2)
                    sst += torch.sum((classes.view(-1) - val_mean) ** 2)
                    
            
            mse = sse / n_samples
            r2 = 1 - sse / sst
            
            if (r2 > self.best_score):
                self.best_model = copy.deepcopy(Model)
                self.best_score = r2

            self.validation_loss.append(mse)
            print(f"epoch: {epoch}")
            print(f" {dt.datetime.now() - start} сек")
            print(f"Loss_val MSE: {mse}")
            print(f"Loss_val R2: {r2}")
            print(f"Loss_train: {loss_tr.item() / len(train_dl)}")
```
